## Removendo dados

### Deletando colunas V
Com base na análise inicial, decidi remover as colunas que começam com 'V'. Essas colunas apresentaram uma quantidade significativa de valores ausentes, o que poderia comprometer a qualidade do modelo. A remoção foi considerada uma abordagem adequada para reduzir a dimensionalidade e melhorar a performance. O código abaixo realiza essa operação

```python
# Removendo as colunas que começam com 'v'
df = df.drop(df.filter(regex='^V').columns, axis=1)
```

###Trocando nan por 0




```python
df = df.fillna(0)
```

### Aplicando label encoder em colunas categoricas

```python
# Converter as colunas que contêm tipos mistos para string
df['P_emaildomain'] = df['P_emaildomain'].astype(str)
df['R_emaildomain'] = df['R_emaildomain'].astype(str)

# Verificar se a conversão foi feita corretamente
print(df[['P_emaildomain', 'R_emaildomain']].dtypes)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

colunas_para_codificar = ['card4', 'card6', 'P_emaildomain', 'R_emaildomain', 'ProductCD']

# Inicializando o LabelEncoder
label_encoder = LabelEncoder()

# Converter todos os valores das colunas para string
for col in colunas_para_codificar:
    df[col] = df[col].astype(str)


# Aplicando o LabelEncoder nas colunas selecionadas
for coluna in colunas_para_codificar:
    df[coluna] = label_encoder.fit_transform(df[coluna])
```

```python
# Converter TransactionDT para o formato de timestamp
df['TransactionDT'] = pd.to_datetime(df['TransactionDT'], unit='s')

# Criar uma nova coluna com a diferença em segundos desde a primeira transação
df['TransactionDT'] = (df['TransactionDT'] - df['TransactionDT'].min()).dt.total_seconds()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Selecionar apenas colunas float, incluindo 'TransactionDT' para temporalidade
df_floats = df.select_dtypes(include=['float64'])

# Adicionar 'TransactionDT' (temporais)
df_floats['TransactionDT'] = df['TransactionDT']

# Adicionar a coluna alvo 'isFraud'
df_floats['isFraud'] = df['isFraud']
```

```python
# Normalizar os dados
scaler = MinMaxScaler()
df_floats_scaled = scaler.fit_transform(df_floats.drop(columns=['isFraud']))

# Separar X e y
X = df_floats_scaled
y = df_floats['isFraud'].values
```

## Criação do Modelo LSTM

```python
# 2. Criar Sequências para o LSTM
def create_sequences(data, target, window_size):
    X_seq, y_seq = [], []
    for i in range(len(data) - window_size):
        X_seq.append(data[i:i + window_size])
        y_seq.append(target[i + window_size])
    return np.array(X_seq), np.array(y_seq)

window_size = 10  # Tamanho da janela de sequência
X_seq, y_seq = create_sequences(X, y, window_size)
```

```python
# 3. Construção do Modelo LSTM
model = Sequential()
model.add(LSTM(50, input_shape=(window_size, X_seq.shape[2]), return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 4. Treinamento do Modelo
model.fit(X_seq, y_seq, epochs=5, batch_size=64, validation_split=0.2)
```
