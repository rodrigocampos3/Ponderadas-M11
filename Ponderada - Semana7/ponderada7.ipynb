```python
# Converter TransactionDT para o formato de timestamp
df['TransactionDT'] = pd.to_datetime(df['TransactionDT'], unit='s')

# Criar uma nova coluna com a diferença em segundos desde a primeira transação
df['TransactionDT'] = (df['TransactionDT'] - df['TransactionDT'].min()).dt.total_seconds()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Selecionar apenas colunas float, incluindo 'TransactionDT' para temporalidade
df_floats = df.select_dtypes(include=['float64'])

# Adicionar 'TransactionDT' (temporais)
df_floats['TransactionDT'] = df['TransactionDT']

# Adicionar a coluna alvo 'isFraud'
df_floats['isFraud'] = df['isFraud']
```

```python
# Normaliza os dados utilizando o MinMaxScaler para transformar as características numéricas para um intervalo entre 0 e 1.
scaler = MinMaxScaler()
df_floats_scaled = scaler.fit_transform(df_floats.drop(columns=['isFraud']))

# Separa as variáveis independentes (X) e a variável dependente (y).
X = df_floats_scaled  # Variáveis independentes: dados normalizados
y = df_floats['isFraud'].values  # Variável dependente: indica se a transação é fraudulenta (1) ou não (0)
```

# Criação do Modelo LSTM

```python
# 2. Criar Sequências para o LSTM
def create_sequences(data, target, window_size):
    X_seq, y_seq = [], []
    for i in range(len(data) - window_size):
        X_seq.append(data[i:i + window_size])
        y_seq.append(target[i + window_size])
    return np.array(X_seq), np.array(y_seq)

window_size = 10  # Tamanho da janela de sequência
X_seq, y_seq = create_sequences(X, y, window_size)
```

###Divisão de treino e teste

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_seq, y_seq, test_size=0.2, random_state=42)
```

Considerando que para a LSTM os dados possuem mais dimensões, precisamos fazer um "achatamento" para transformá-los de um formato tridimensional (n_samples, timesteps, n_features) para um formato bidimensional (n_samples, timesteps * n_features). Esse processo é necessário porque muitas técnicas de pré-processamento de dados, como o undersampling, requerem que os dados estejam em duas dimensões.

Depois de aplicar o undersampling, que tem o objetivo de equilibrar a distribuição das classes no conjunto de treinamento, precisamos reformatar os dados de volta para três dimensões. Isso é feito ao reorganizar os dados achatados de modo que cada amostra retorne ao seu formato original, mantendo a estrutura temporal e as características necessárias para a LST

###Definindo modelo e hiperparametros

O modelo inicia com uma camada LSTM com 50 unidades, recebendo entradas com uma forma específica definida pelo tamanho da janela (window_size) e o número de recursos disponíveis em X_seq. Esta camada retorna sequências, permitindo que a informação temporal seja mantida. Em seguida, adiciona-se uma segunda camada LSTM, também com 50 unidades, que processa as saídas da camada anterior. Para a saída do modelo, uma camada densa é incorporada com uma função de ativação sigmoide, destinada a prever uma única saída binária. Por fim, o modelo é compilado utilizando o otimizador Adam e a função de perda de entropia cruzada binária, além de monitorar a métrica de precisão durante o treinamento.

```python
# 3. Construção do Modelo LSTM
model = Sequential()
model.add(LSTM(50, input_shape=(window_size, X_seq.shape[2]), return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

# Treinamento do Modelo

```python
# 4. Treinamento do Modelo
history = model.fit(X_train_res, y_train_res, epochs=5, batch_size=64, validation_split=0.2)
```

```python
from sklearn.metrics import classification_report, roc_auc_score

# Faça previsões
y_pred = (model.predict(X_test) > 0.5).astype("int32")

# Imprima o relatório de classificação (precisão, recall, F1-score)
print(classification_report(y_test, y_pred))

# AUC-ROC
y_pred_prob = model.predict(X_test).ravel()
auc_roc = roc_auc_score(y_test, y_pred_prob)
print(f'AUC-ROC: {auc_roc}')
```
